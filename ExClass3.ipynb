```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqUtils import MeltingTemp, molecular_weight, GC
from matplotlib import pyplot as plt
```

```python
def load_fasta(filename):
    record = SeqIO.read(filename, 'fasta')
    return record
```

```python
record = load_fasta('sequence.fasta')
print(record)

seqA = record.seq[:round(len(record.seq)/2)]
seqB = record.seq[round(len(record.seq)/2):]
print(seqA)
print(seqB)
```

```python
len_seqA = len(seqA)
len_seqB = len(seqB)
print(len_seqA)
print(len_seqB)
```

```python
count_seqA = seqA.count('CGC')
count_seqB = seqB.count('CGC')
print(count_seqA)
print(count_seqB)
```

```python
find_seqA = seqA.find('CAGTC')
find_seqB = seqB.find('CAGTC')
print(find_seqA)
print(find_seqB)
```

```python
seqC = seqA[:15] + seqB[10:]
print(seqC)

print(seqC[::-1])
```

```python
mw_seqA = molecular_weight(seqA)
mw_seqB = molecular_weight(seqB)
print(mw_seqA)
print(mw_seqB)
```

```python
mt_seqA = MeltingTemp.Tm_Wallace(seqA)
mt_seqB = MeltingTemp.Tm_Wallace(seqB)
print(mt_seqA)
print(mt_seqB)
```

```python
gc_seqA = GC(seqA)
gc_seqB = GC(seqB)
print(gc_seqA)
print(gc_seqB)
```

```python
at_seqA = 100 - gc_seqA
at_seqB = 100 - gc_seqB
print(at_seqA)
print(at_seqB)
```

```python
fig, axis = plt.subplots(2,2)

# GC
axis[0][0].bar(['seqA', 'seqB'], [gc_seqA, gc_seqB])
axis[0][0].set_title("GC Content")

# AT
axis[0][1].bar(['seqA', 'seqB'], [at_seqA, at_seqB])
axis[0][1].set_title("AT Content")

# Melting Point
axis[1][0].bar(['seqA', 'seqB'], [mt_seqA, mt_seqB])
axis[1][0].set_title("Melting Point")

# Molecular Weight
axis[1][1].bar(['seqA', 'seqB'], [mw_seqA, mw_seqB])
axis[1][1].set_title("Molecular Weight")

fig.tight_layout()
fig.show()
```
